```python
from summa.summarizer import summarize
import os
```

```python
valid_dir = 'C:/My folders/Deva paul/AI/SEM 5/356 NLP/NLP Project/New folder/FNS2023_Datasets/English/validation'
annual_reports_dir = "annual_reports"
```

```python
file_id = '31794'
with open(os.path.join(valid_dir,annual_reports_dir, str(file_id)+'.txt'), "r", encoding='utf-8') as report_file:
    doc = report_file.read()
print(doc)
```

```python
num_file = 0
for file in os.listdir(os.path.join(valid_dir, annual_reports_dir)):
    try:
        print("Processing File Number: ", num_file)
        num_file = num_file +1 
        file_id = file.split('.')[0]
   
        with open(os.path.join(valid_dir,annual_reports_dir, str(file_id)+'.txt'), "r", encoding='utf-8') as report_file:
            doc = report_file.read()
            summary = summarize(doc, words=1000)
            
        with open('english/'+ file_id+'IT-356'+'.txt', 'w', encoding='utf-8') as f:
            f.write(str(summary))
        
        if ".DS_Store" in file:
            continue
    except Exception as e:
        print(file, e)
```
